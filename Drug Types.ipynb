```python
import csv 
import pandas as pd

def findCategories(root, drugbank_id, list_cats):

    for child in root: #go through every child node
        for tag in child.findall('{http://www.drugbank.ca}categories'): #finds the categories tag
            for category in tag: #goes through first category tag
                for subcategory in category: #goes through second category tag
                    if subcategory.tag == '{http://www.drugbank.ca}category': #checks the right namespace
                        for tag in child.findall('{http://www.drugbank.ca}drugbank-id'): #searches for drugbank id
                                if 'primary' in tag.attrib: #gives only primary id 
                                    if drugbank_id == tag.text: #checks for the right drug id
                                        if "Anticoag" in subcategory.text: #the rest of the if statements check if they are cardiovascular drugs then adds the category to a list
                                            list_cats.append(subcategory.text)
                                            break
                                        if "Vasodilat" in subcategory.text:
                                            list_cats.append(subcategory.text)
                                            break
                                        if "Antiplatelet" in subcategory.text:
                                            list_cats.append(subcategory.text)
                                            break
                                        if "Angioten" in subcategory.text:
                                            list_cats.append(subcategory.text)
                                            break
                                        if "Beta Block" in subcategory.text:
                                            list_cats.append(subcategory.text)
                                            break
                                        if "Calcium Channel" in subcategory.text:
                                            list_cats.append(subcategory.text)
                                            break
                                        if "ACE" in subcategory.text:
                                            list_cats.append(subcategory.text)
                                            break
                                        if "Cholesterol" in subcategory.text:
                                            list_cats.append(subcategory.text)
                                            break
                                        if "Digitalis" in subcategory.text:
                                            list_cats.append(subcategory.text)
                                            break
                                        if "Diuretic" in subcategory.text:
                                            list_cats.append(subcategory.text)
                                            break
    

#creating a data frame to then import to csv
data = []  # will contain a list which includes [drugbank id, [categories]]  
for drug in common_cardio_drugs:
    list_cats = []
    findCategories(root, drug, list_cats)
    data.append([drug, list_cats])
df = pd.DataFrame(data, columns = ['Drugbank ID', 'Categories'])   


df.to_csv(r'C:\Users\aarus\Downloads\node_list.csv')

                                       
                                            
                                      
```
